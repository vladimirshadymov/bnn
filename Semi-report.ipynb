```python
def hard_tanh(x):
    arr = []
    for elem in x:
        if np.abs(elem)>1: arr.append(np.sign(elem))
        else: arr.append(elem)
    return arr
```

```python
x_value = np.linspace(-3, 3, 100000)
x_dots = np.random.rand(10)*2-1.
plt.plot(x_value, np.sign(x_value), label=r'Forward')
plt.plot(x_value, hard_tanh(x_value), label=r"Backprop")
plt.grid(True)
plt.ylim(-2, 2)
plt.scatter(x_dots, np.zeros_like(x_dots), label=r'$W$')
plt.scatter(x_dots, np.sign(x_dots), label=r'$W^b$')
plt.legend(loc='best')
plt.xlabel('Weights')
plt.ylabel("Activation")
plt.savefig("activation.png", dpi=300, transparent=True)
```
