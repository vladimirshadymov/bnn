```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
svhn_cnn = pd.read_csv("model/svhn_cnn_report.csv")
svhn_cnn_pruning = pd.read_csv("model/svhn_cnn_report_pruning_mode.csv")
svhn_bnn = pd.read_csv("model/svhn_conv_bnn_report.csv")
svhn_bnn_pruning = pd.read_csv("model/svhn_conv_bnn_pruning_mode_report.csv")
```

```python
plt.plot(svhn_cnn["Test accuracy"][0:251], label="CNN")
plt.plot(svhn_cnn_pruning["Test accuracy"][0:251], label="CNN Pruning")
plt.plot(svhn_bnn["Test accuracy"][0:251], label="BNN")
plt.plot(svhn_bnn_pruning["Test accuracy"][0:251], label="BNN Pruning")
plt.grid(True)
plt.legend()
plt.title("SVHN Dataset")
plt.xlabel("Epoch")
plt.ylabel("Accuracy, %")
plt.ylim(80,)
plt.savefig("svhn.png", dpi=300, transparent=True)
```

```python
cifar10_cnn = pd.read_csv("model/cifar10_cnn_report.csv")
cifar10_cnn_pruning = pd.read_csv("model/cifar10_cnn_report_pruning_mode.csv")
cifar10_bnn = pd.read_csv("model/cifar10_conv_bnn_report.csv")
cifar10_bnn_pruning = pd.read_csv("model/cifar10_conv_bnn_pruning_mode_report.csv")
```
